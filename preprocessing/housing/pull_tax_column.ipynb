```python
unique_addresses = df_utilities['Premise Address'].unique()
unique_addresses = pd.Series(unique_addresses)
unique_addresses.head()
```

```python
df_taxes_list = []
path = '~/Dropbox/CDS-2019-AlbanyHub/Raw-Data/'
for i in range(2009, 2020):
    df_taxes_list.append(pd.read_csv(path+'AY'+str(i)+ ' Real Property_Personal Property.csv'))
df_full_taxes = pd.concat(df_taxes_list, ignore_index=True)
df_full_taxes.head
```

```python
add = []

for (number, direction, name, suffix) in zip(df_full_taxes['HOUSE_NO'], df_full_taxes['STDIRECT'], 
                                          df_full_taxes ['STREET_NAM'], df_full_taxes['STREET_TYPE']):
    word = ''
    if number != 0:
        word += str(number)
    if direction != '  ' and not pd.isna(direction):
        word += ' ' + str(direction).strip()
    word += ' ' + name.strip()
    if suffix != '    ' and not pd.isna(suffix):
        word += ' ' + str(suffix).upper().strip()
    add.append(word)

address = pd.Series(add)
address = pd.Series(address.unique())
address.shape
```

```python
unique_addresses.shape
```

```python
mismatched = unique_addresses[~unique_addresses.isin(address)]
mismatched.shape
```

```python
def lev_list(address_list, address,i):
    temp_list = []
    temp_list_u = []
    for u_address in address_list:
        if Levenshtein.distance(address, u_address.upper()) <= i:
            temp_list.append(address)
            temp_list_u.append(u_address.upper())
    temp_list_series = pd.Series(temp_list)
    temp_list_u_series = pd.Series(temp_list_u)
    return temp_list_u_series
```

```python
def gen_list_none(address_list, mismatched, i):
    tmp_list = []
    counter = 0
    for address in mismatched:
        if counter % 100 == 0:
            print(str(counter))
        if lev_list(address_list, address, i).shape[0] == 0:
            tmp_list.append(address)
        counter += 1
    return pd.Series(tmp_list)
```
