```python
import pandas as pd
```

```python
df = pd.read_csv('~/Dropbox (Amherst College)/CDS-2019-AlbanyHub/blockgroups_first_100.csv')
```

```python
df.head()
```

```python
s_lat_long = df['addrlocat']
```

```python
lat = []
long = []
for address in s_lat_long:
    word = address.split(',')
    lat.append(word[0])
    long.append(word[1])
```

```python
lat = pd.Series(lat)
long = pd.Series(long)
```

```python
df2 = pd.concat([lat, long, pd.Series(["GA"]*100)], axis = 1)
```

```python
df2.columns = ["Latitude", "Longitude", "State"]
```
