```python
# Examine the junction table and look at addresses that were geocoded as being in the same location
# Which could indicate that there was a small typo
jt = pd.read_csv('/home/mirabel/Dropbox (GaTech)/CDS-2019-AlbanyHub/ToDatabase/addr_junct_table.csv')
jt['xy'] = [str(jt.loc[i,'Xcoord'])+" "+str(jt.loc[i,'Ycoord']) for i in range(len(jt))]
group = jt.groupby(by='xy').count()
xy_list = group.loc[group['Address']>1].index
repeated = jt[jt['xy'].isin(xy_list)]
```
