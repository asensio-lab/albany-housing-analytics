```python
def duplicated(df):
    cols = list(df)
    uniques = []
    for col in cols:
        valuelist = df[col].value_counts()
        if len(valuelist) == 0:
            uniques.append(True)
        else:
            uniques.append(valuelist.iloc[0]==1)
    return uniques
```

```python
def construct_summary(df):
    return pd.DataFrame({'Datatype': df.dtypes, 'Nulls': df.isna().any(),
             'Uniques': duplicated(df)})
```

```python
def sql_helper(summary):
    lines = []
    for i in range(len(summary)):
        col_name = summary.index[i]
        datatype = str(summary.iloc[i]['Datatype'])
        nulls = summary.iloc[i]['Nulls']
        line = ""
        line += '[' + col_name +'] '
        if datatype == 'object':
            line += "[VARCHAR](255) "
        elif datatype == 'int64':
            line += "[INT] "
        elif datatype == 'float64':
            line += "[FLOAT] "
        else:
            line += "Inspect " + str(i) + ": " + col_name
        if nulls:
            line += "NOT NULL,"
        else:
            line += "NULL,"
        lines.append(line)
    return lines
```

```python
def sql_helper_write(lines, name):
    with open(name+'_sql_helper.txt', 'w') as f:
        for line in lines:
            f.write(line+'\n')
```

```python
lines = sql_helper(block_summary)
sql_helper_write(lines, 'block_group')
```

```python
df_tract = df_tract.replace({'(X)':np.nan, '-':np.nan, 'N':np.nan})
```

```python
def float_convert(convert): #helper to return list of columns that could be converted to floats
    can_be_float = True
    for i in range(len(convert)):
        try:
            number = float(convert.iloc[i])
        except:
            can_be_float = False
    return can_be_float
```

```python
def int_convert(convert): #helper to return list of columns that could be converted to int
    can_be_int = True
    for i in range(len(convert)):
        try:
            number = float(convert.iloc[i])
            if math.isnan(number):
                continue
            if number - int(number) != 0.0:
                can_be_int = False
        except:
            can_be_int = False
    return can_be_int
```

```python
each_row_float = [] #wildly inconsistent but whatever
for i in range(len(list(df_tract))):
    each_row_float.append(float_convert(df_tract.iloc[:,i]))
each_row_float = pd.Series(each_row_float)
```

```python
each_row_int = [] #wildly inconsistent but whatever
for i in range(len(list(df_tract))):
    each_row_int.append(int_convert(df_tract.iloc[:,i]))
each_row_int = pd.Series(each_row_int)
```

```python
each_row_float.index = list(df_tract)
each_row_int.index = list(df_tract)
```
