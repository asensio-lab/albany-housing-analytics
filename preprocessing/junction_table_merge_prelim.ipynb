```python
df_housing_projects.drop(['City', 'State', 'XY Coordinates'], inplace = True, axis = 1)
```

```python
df_junction_table.columns = ['PrimaryID', 'Address', 'Xcoord', 'Ycoord', 'Tract', 'BlockGroup']
```

```python
s_housing_projects = df_housing_projects['Address']
s_junction_table = df_junction_table['Address']
s_primaryid = df_junction_table['PrimaryID']
```

```python
primaryid_list = []
notfound_list = []
for house in s_housing_projects:
    try:
        index = s_junction_table[s_junction_table == house].index[0]
        primaryid_list.append(s_primaryid[index])
    except:
        primaryid_list.append("NA")
        notfound_list.append(house)
primaryID = pd.Series(primaryid_list)
notFound = pd.Series(notfound_list)
```

```python
primaryID.shape
```

```python
df_housing_projects['PrimaryID'] = primaryID
```
