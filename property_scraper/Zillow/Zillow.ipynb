# This file investigates the zillow api

```python
import json
import requests
import pandas as pd
from xml.etree import ElementTree
path = '~/Dropbox/CDS-2019-AlbanyHub/'
```

## A single query

```python
zwsid = ""
addr = '1702 W OAKRIDGE DR'
csz = 'ALBANY GA 31707'
rz = True
```

```python
params = {'zws-id':zwsid, 'address':addr, 'citystatezip':csz, 'rentzestimate':rz}

response = requests.get("http://www.zillow.com/webservice/GetDeepSearchResults.htm", params=params)
```

# Parse response for content

```python
#Content is in xml format -> turn into a dict
response
my_dict = {}
tree = ElementTree.fromstring(response.content)
for child in tree.iter():
    my_dict[child.tag] = child.text
```

```python
# Print all response content
for key, value in my_dict.items():
    print(key, value)
my_dict.keys()
```

```python
#Remove some content that isn't useful
k = list(my_dict.keys())
k = k[10:-3]
k[1:6] = []
my_dict = {a:my_dict[a] for a in k}
```

```python
for j in my_dict.keys():
    print(j, my_dict[j]) 
```

```python
all_addresses = pd.read_csv(path+'Processed-Data/all_addresses.csv')
all_addresses.head()
```

```python
df = pd.DataFrame(data=my_dict, index=[0])
df.head()
76421669
```

```python
#Query 100 addresses
for i in range(1,100):
    #get address number i
    addr = all_addresses.loc[i, 'line1']
    csz = all_addresses.loc[i, 'line2']
    params = {'zws-id':zwsid, 'address':addr, 'citystatezip':csz, 'rentzestimate':rz}
    #Query zillow for the deep search results of this location
    response = requests.get("http://www.zillow.com/webservice/GetDeepSearchResults.htm", params=params)
    #add the query to a dictionary
    my_dict = {}
    tree = ElementTree.fromstring(response.content)
    for child in tree.iter():
        my_dict[child.tag] = child.text
    #Drop irrelevant keys
    k = list(my_dict.keys())
    k = k[10:-3]
    k[1:6] = []
    my_dict = {a:my_dict[a] for a in k}
    df = df.append(my_dict, ignore_index=True)
    
```
