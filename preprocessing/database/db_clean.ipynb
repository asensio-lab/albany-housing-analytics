## Linking all of the ids

```python
def linker(df_modified, df_info, info_key_name, info_value_name, modified_col, not_found=None): # modified from map_addr_key
    id_dict = df_info.set_index(info_key_name).to_dict()[info_value_name]
    s_modified = df_modified[modified_col]
    linked_ids = [None] * len(df_modified)
    not_found_dict = {}
    counter = 0
    for i in range(len(s_modified)):
        if counter % 100000 == 0:
            print(str(counter))
        try:
            linked_ids[i] = id_dict[s_modified[i]]
        except KeyError:
            if not_found == None:
                print("FATAL ERROR. Corresponding key not found.")
            linked_ids[i] = id_dict[not_found]
            not_found_dict[i] = s_modified[i]
        counter += 1
    # Change this in the future to have linked ids be a series
    return linked_ids, pd.DataFrame(data={'key':list(not_found_dict.values()), 
                                          'location': list(not_found_dict.keys())})
```
