# vacancy

```python
new = vacancy["GEO.display-label"].str.split(", ", expand = True)   
vacancy["block"] = new[0] 
vacancy["tract"] = new[1]
vacancy["county"] = new[2]
```

```python
vacancy = vacancy.drop('GEO.display-label', axis = 1)
```

```python
l = len(vacancy['tract'])
vacancy1 = vacancy.copy()
re_string = 'Census Tract (\d+(?:.\d+)?)'
for i in range(1,l):
    s = vacancy.loc[i, 'tract']
    m = re.search(re_string, s)
    vacancy1.loc[i, 'tract'] = m.group(1)
```

```python
l = len(vacancy['block'])
re_string = 'Block Group (\d+(?:.\d+)?)'
for i in range(1,l):
    s = vacancy.loc[i, 'block']
    m = re.search(re_string, s)
    vacancy1.loc[i, 'block'] = m.group(1)
```

```python
tract_series = vacancy1['tract']
l = len(vacancy1['tract'])
for i in range(1,l):
    if '.' in tract_series[i]:
        s = tract_series[i].split('.')
        if len(s[0]) == 1:
            tract_series[i] = s[0] + s[1]
        elif len(s[0]) == 2:
            tract_series[i] = s[0] + s[1]
        elif len(s[0]) == 3:
            tract_series[i] = s[0] + s[1]
    else:
        if len(tract_series[i]) == 3:
            tract_series[i] = vacancy1['tract'][i] + '00'
        elif len(tract_series[i]) == 2:
            tract_series[i] = vacancy1['tract'][i] + '00'
        elif len(tract_series[i]) == 1:
            tract_series[i] = vacancy1['tract'][i] + '00'
        elif len(tract_series[i]) == 4:
            tract_series[i] = vacancy1['tract'][i] + '00'
```
