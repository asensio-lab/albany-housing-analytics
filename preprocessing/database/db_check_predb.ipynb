```python
def duplicated(df):
    cols = list(df)
    uniques = []
    for col in cols:
        valuelist = df[col].value_counts()
        if len(valuelist) == 0:
            uniques.append(True)
        else:
            uniques.append(valuelist.iloc[0]==1)
    return uniques
```

```python
def construct_summary(df):
    return pd.DataFrame({'Datatype': df.dtypes, 'Nulls': df.isna().any(),
             'Uniques': duplicated(df)})
```

```python
address_summary = construct_summary(df_address)
utility_summary = construct_summary(df_utility)
project_summary = construct_summary(df_project)
real_summary = construct_summary(df_real)
blockgroup_summary = construct_summary(df_blockgroup)
tract_summary = construct_summary(df_tract)
tax_summary = construct_summary(df_tax)
weather_summary = construct_summary(df_weather)
```
