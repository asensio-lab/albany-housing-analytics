```python
import numpy as np
import pandas as pd
```

```python
infile = '~/Dropbox (GaTech)/CDS-2019-AlbanyHub/Raw-Data/Housing/All CED Funded Projects 2007-2017.xlsx'
xlsx = pd.ExcelFile(infile)
df1 = pd.read_excel(xlsx, 'HOME Funded Projects')
df2 = pd.read_excel(xlsx, 'CDBG Funded Projects')
```

```python
df = pd.concat([df1, df2])
df.index = range(len(df))
df.head()
```

```python
df1.to_csv('HOME_Funded_Projects.csv', index = False)
df2.to_csv('CDBG_Funded_Projects.csv', index = False)
df.to_csv('Total_Funded_Projects.csv', index = False)
```
