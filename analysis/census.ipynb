```python
import pandas as pd
```

```python
block_group = pd.read_csv('/Users/davidreynolds/Dropbox (GaTech)/CDS-2019-AlbanyHub/ToDatabase/full_blockgroup.csv')
tract = pd.read_csv('/Users/davidreynolds/Dropbox (GaTech)/CDS-2019-AlbanyHub/ToDatabase/full_tract.csv')
```

```python
#Remove All Observations That Aren't in Dougherty County
block_group = block_group[block_group.county == 'Dougherty County']
```

```python
#Remove Irrelevant Columns
block_group = block_group.drop(['geo_id', 'geo_id2', 'county', 'block_tract'], axis = 1)
```

```python
#Create Pct Fields
block_group["in_laborforce_pct"] = block_group["total_in_laborforce"] / block_group["total_labor"] * 100
block_group["notin_laborforce_pct"] = block_group["total_notin_laborforce"] / block_group["total_labor"] *100
block_group["emp_civil_labor_pct"] = block_group["emp_civil_labor"] / block_group["total_civil_labor"] * 100
block_group["unemp_civil_labor_pct"] = block_group["unemp_civil_labor"] / block_group["total_civil_labor"] * 100
block_group["occ_homes_pct"] = block_group["total_occ_homes"] / block_group["total_vacdata"] * 100
block_group["vac_homes_pct"] = block_group["total_vac_homes"] / block_group["total_vacdata"] * 100
block_group["owner_pct"]  = block_group["total_owner"] / block_group["total_ordata"] * 100
block_group["renter_pct"] = block_group["total_renter"] / block_group["total_ordata"] * 100
```

```python
#Reorder Columns
cols = list(block_group)
cols.insert(0, cols.pop(cols.index('')))
block_group = block_group.loc[:, cols]
```
