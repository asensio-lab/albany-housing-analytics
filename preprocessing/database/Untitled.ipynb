```python
def create_dict():
    dic = {}
    for i in range(len(df_real)):
        record = df_real.iloc[i]
        size = record['Size']
        ids = record['AddressID']
        dic[ids] = size
    return dic
```

```python
dic = create_dict()
```

```python
tmp = []
for i in range(len(df_avg)):
    if i % 10000 == 0:
        print(str(i))
    record = df_avg.iloc[i]['AddressID']
    try:
        tmp.append(dic[record])
    except KeyError:
        tmp.append("NA")
tmp = pd.Series(tmp)
```

```python
df_avg['Normalized'] = df_avg['AvgCons'] / tmp
```

```python
monthsdays = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
```

```python
def returnDays(yearmonth):
    spl = yearmonth.split("_")
    year = int(spl[0])
    month = int(spl[1])
    if month == 2:
        if year == 2012 or year == 2016:
            return 29
    return monthsdays[month]
```

```python
ym = df_avg['YearMonth'].apply(returnDays)
```
